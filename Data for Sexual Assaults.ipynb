```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from statistics import mean
from statistics import stdev
import numpy as np
```

```python
sexual_offenders = pd.read_csv('sexoffenders.csv').fillna(0)
```

```python
# Replace the data error with 'Oct' to '10'
sexual_offenders['AgeGroup'] = sexual_offenders.AgeGroup.replace('Oct-14', '10-14')

# Readability
sexual_offenders['AgeGroup'] = sexual_offenders.AgeGroup.replace('65 years and over', '65+')
```

```python
# Convert to int()
sexual_offenders.y_2012 = sexual_offenders.y_2012.astype(int)
sexual_offenders.y_2013 = sexual_offenders.y_2013.astype(int)
sexual_offenders.y_2014 = sexual_offenders.y_2014.astype(int)
sexual_offenders.y_2015 = sexual_offenders.y_2015.astype(int)
sexual_offenders.y_2016 = sexual_offenders.y_2016.astype(int)
```

```python
# Line Plot for Male Sex Offenders (Number of Assaults vs Age)
plt.plot(sexual_offenders['AgeGroup'][:12],sexual_offenders['y_2012'][:12], c = 'blue')
plt.plot(sexual_offenders['AgeGroup'][:12],sexual_offenders['y_2013'][:12], c = 'orange')
plt.plot(sexual_offenders['AgeGroup'][:12],sexual_offenders['y_2014'][:12], c = 'green')
plt.plot(sexual_offenders['AgeGroup'][:12],sexual_offenders['y_2015'][:12], c = 'red')
plt.plot(sexual_offenders['AgeGroup'][:12],sexual_offenders['y_2016'][:12], c = 'purple')

plt.legend(('2012', '2013', '2014', '2015', '2016'))
plt.title('Male Sex Offenders in 2012-2016 by Age Groups')
plt.ylabel('Recorded number of Male Offenders')
plt.xlabel('Age Groups')
plt.xticks(rotation=70)

plt.show()
```

```python
# Line Plot for Female Sex Offenders (Number of Assaults vs Age)
plt.plot(sexual_offenders['AgeGroup'][12:],sexual_offenders['y_2012'][12:], c = 'blue')
plt.plot(sexual_offenders['AgeGroup'][12:],sexual_offenders['y_2013'][12:], c = 'orange')
plt.plot(sexual_offenders['AgeGroup'][12:],sexual_offenders['y_2014'][12:], c = 'green')
plt.plot(sexual_offenders['AgeGroup'][12:],sexual_offenders['y_2015'][12:], c = 'red')
plt.plot(sexual_offenders['AgeGroup'][12:],sexual_offenders['y_2016'][12:], c = 'purple')

plt.legend(('2012', '2013', '2014', '2015', '2016'))
plt.title('Female Sex Offenders in 2012-2016 by Age Groups')
plt.ylabel('Recorded number of Female Offenders')
plt.xlabel('Age Groups')
plt.xticks(rotation=70)

plt.show()
```

```python
# Add a new column indicating the average number of offenders according to age groups over the 5 years
sexual_offenders['Average'] = sexual_offenders.mean(axis=1)

# Standard Deviations
so_m_stdev = sexual_offenders.std(axis=1)[:12]
so_f_stdev = sexual_offenders.std(axis=1)[12:]
```

```python
plt.clf()

# Bar Plot of Average Male and Female Offenders according to Age Group
plt.bar(sexual_offenders['AgeGroup'][:12], sexual_offenders['Average'][:12], yerr = so_m_stdev, color = 'royalblue')
plt.bar(sexual_offenders['AgeGroup'][12:], sexual_offenders['Average'][12:], yerr = so_f_stdev, color = 'fuchsia')

plt.legend(('Male', 'Female'))
plt.title('Sexual Offenders by Gender')
plt.ylabel('Total number of cases reported')
plt.xlabel('Age Groups')
plt.xticks(rotation=70)

plt.show()
```
