```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
from statistics import mean
from statistics import stdev
import numpy as np
import string
```

```python
data = pd.read_csv('CrimebyCategoryxgender.csv', encoding='latin1').fillna(0)
```

```python
# Add a new column indicating the average number of cases according to age groups over the 5 years
data['Average'] = data.mean(axis=1)

# Years
years = ['2012','2013', '2014', '2015', '2016']
```

```python
# Convert to int()
data.y_2012 = data.y_2012.astype(int)
data.y_2013 = data.y_2013.astype(int)
data.y_2014 = data.y_2014.astype(int)
data.y_2015 = data.y_2015.astype(int)
data.y_2016 = data.y_2016.astype(int)
data.Average = data.Average.astype(int)
```

```python
# Cases of Crimes against the person
cat_A = data[data.cat.str.contains('A')]
crimes_against_the_person = {'A10': 'Homocide and related offences', 'A20': 'Assault and related offences',
                            'A30': 'Sexual offences', 'A40': 'Abduction and related offences',
                             'A50': 'Robbery', 'A60': 'Blackmail and extortion',
                             'A70': 'Stalking, harassment and threatning behaviour',
                            'A80': 'Dangerous and negligent acts endangering people'}
```

```python
# Line Plot of Male offenders for Crimes against the person
plt.clf()

for i in range(8):
    plt.plot(years, cat_A.loc[i][2:-1].astype(int))

plt.legend([crimes_against_the_person[i] for i in cat_A.cat[8:]], loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Cases of Crimes against the person (Male)')
plt.ylabel('Number of cases reported')
plt.xlabel('Years')
plt.yscale('log')
plt.yticks([10, 100, 1000, 10000, 30000], ['10', '100', '1000', '10000', '30000'])
plt.grid()

plt.show()
```

```python
# Line Plot of Female offenders for Crimes against the person
plt.clf()

for i in range(8, 16):
    plt.plot(years, cat_A.iloc[i][2:-1].astype(int))

plt.legend([crimes_against_the_person[i] for i in cat_A.cat[8:]], loc='center left', bbox_to_anchor=(1, 0.5))
plt.title('Cases of Crimes against the person (Female)')
plt.ylabel('Number of cases reported')
plt.xlabel('Years')
plt.yscale('log')
plt.yticks([10, 100, 1000, 10000], ['10', '100', '1000', '10000'])
plt.grid()

plt.show()
```

```python
plt.clf()

# Standard Deviations
m_stdev = cat_A.std(axis=1)[:8]
f_stdev = cat_A.std(axis=1)[8:]

# Bar Plot of Average Male and Female Offenders according to Age Group
plt.bar(cat_A.cat[:8], cat_A.Average[:8], yerr = m_stdev, color = 'royalblue')
plt.bar(cat_A.cat[8:], cat_A.Average[8:], yerr = f_stdev, color = 'fuchsia')

plt.legend(('Male', 'Female'))
plt.title('Cases of Crimes against the person by Gender')
plt.ylabel('Number of cases reported')
plt.xticks(cat_A.cat.loc[:8], [crimes_against_the_person[i] for i in cat_A.cat[8:]], rotation=90)

plt.show()
```
